```python
# Import the library we need.
import numpy as np
import itertools
import json
import jieba
from gensim.summarization import bm25
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter
```

```python
def jiebaCut(cpuName):
    query = []
    ss = list(jieba.cut(cpuName, cut_all = False))
    for s in ss: 
        if s != " " and s != "_" and s != "-" and s != "" and s != "（" and s != "）"  and s != "\n":
            if s == "+" :
                query.append("Plus")
            else: 
                query.append(s)
    #print(query)    
    return query
```

```python
corpus = []
for cpuName in dScore[0]:
    corpus.append(jiebaCut(cpuName))
```

```python
idQuery = list(df["Unnamed: 0"])
cpuQuery = list(df["CPU型号"])
priceQuery = list(df["价格"])

weightQuery = list(df["机身重量"])
for i in range(len(weightQuery)): 
    weightQuery[i] = float(weightQuery[i].split("g")[0])
    
romQuery = list(df["机身内存"])
for i in range(len(romQuery)): 
    if romQuery[i] == '1TB':
        romQuery[i] = float('1024')
    else :
        romQuery[i] = float("".join(filter(str.isdigit, romQuery[i])))
    
screenQuery = list(df["屏幕尺寸"])
for i in range(len(weightQuery)): 
    screenQuery[i] = float(screenQuery[i].split("英寸")[0])
    
hzQuery = list(df["屏幕刷新率"])
for i in range(len(weightQuery)): 
    hzQuery[i] = float("".join(filter(str.isdigit, hzQuery[i])))

powerQuery = list(df["充电功率"])
for i in range(len(weightQuery)): 
    powerQuery[i] = float("".join(filter(str.isdigit, powerQuery[i])))
    
XpowerQuery = list(df["无线充电"])
for i in range(len(weightQuery)): 
    if XpowerQuery[i] == '不支持':
        XpowerQuery[i] = float('0')
    else:    
        XpowerQuery[i] = float("".join(filter(str.isdigit, XpowerQuery[i])))

osQuery = list(df["系统"])
for i in range(len(weightQuery)): 
    if osQuery[i] == 'iOS':
        osQuery[i] = float('1')
    elif osQuery[i] == 'Android':
        osQuery[i] = float('2')
    else:
        osQuery[i] = float('3')
        
netQuery = list(df["5G网络"])
for i in range(len(weightQuery)): 
    if netQuery[i] == '支持5G':
        netQuery[i] = float('1')
    else:
        netQuery[i] = float('0')

simQuery = list(df["SIM卡数量"])
for i in range(len(weightQuery)): 
    simQuery[i] = float("".join(filter(str.isdigit, simQuery[i])))
    
bcamQuery = list(df["后摄主像素"])
for i in range(len(weightQuery)): 
    bcamQuery[i] = float("".join(filter(str.isdigit, bcamQuery[i])))
    
fcamQuery = list(df["前摄主像素"])
for i in range(len(weightQuery)): 
    fcamQuery[i] = float("".join(filter(str.isdigit, fcamQuery[i])))
```

```python
def Z_Score(itemList, zscoreData, name):
    nameQ = name + "σ"
    nameAve = name + "Ave"
    itemAve = sum(itemList)/len(itemList)   
    itemSum = 0 
    for i in range(len(itemList)):
        itemSum = itemSum + ((itemList[i] - itemAve)*(itemList[i] - itemAve))

    q = np.sqrt(itemSum / len(itemList))
    
    itemScoreList = []
    for i in range(len(itemList)): 
        itemScore = (itemList[i] - itemAve)/q
        itemScoreList.append(itemScore)
        #print("初始：%f  标准化: %f"%(itemList[i], itemScore))
    zscoreData[nameQ] = q
    zscoreData[nameAve] = itemAve
    print("%s 平均值：%f  标准化最大值：%f  最小值: %f"%(name,itemAve, max(itemScoreList), min(itemScoreList)))
    num= 0
    for i in range(len(itemScoreList)): 
        if itemScoreList[i] <= 2 and itemScoreList[i] >= -2:
            num = num + 1
    print("[-2， 2]数值占比: %f"%(num/len(itemScoreList)))
    return itemScoreList
```

```python
#处理CPU，利用BM25匹配
cpuSingle = []
cpuMulti = []
for cpuName in cpuQuery:
    query = jiebaCut(cpuName)
    bm25Model = bm25.BM25(corpus)
    scores = bm25Model.get_scores(query)
    indexNum = scores.index(max(scores))
    print("原始文本： %s +++++++++++++++++  匹配到: %s  Geekbench5跑分: %s"%(cpuName, dScore[0][indexNum], dScore[1][indexNum]))
    cpuSingle.append(float(dScore[1][indexNum].split(" / ")[0]))
    cpuMulti.append(float(dScore[1][indexNum].split(" / ")[1]))
```

```python
zscoreData = {}
priceScoreList = Z_Score(priceQuery, zscoreData, "price")
cpuSingleScoreList = Z_Score(cpuSingle, zscoreData, "cpuSingle")
cpuMultiScoreList = Z_Score(cpuMulti, zscoreData, "cpuMulti")
weightScoreList = Z_Score(weightQuery, zscoreData, "weight")
romScoreList = Z_Score(romQuery, zscoreData, "rom")
screenScoreList = Z_Score(screenQuery, zscoreData, "screen")
hzScoreList = Z_Score(hzQuery, zscoreData, "hz")
powerScoreList = Z_Score(powerQuery, zscoreData, "power")
XpowerScoreList = Z_Score(XpowerQuery, zscoreData, "Xpower")
osScoreList = Z_Score(osQuery, zscoreData, "os")
netScoreList = Z_Score(netQuery, zscoreData, "net")
simScoreList = Z_Score(simQuery, zscoreData, "sim")
bcamScoreList = Z_Score(bcamQuery, zscoreData, "bcam")
fcamScoreList = Z_Score(fcamQuery, zscoreData, "fcam")
```

```python
zscoreDataList = []
for i in range(len(weightQuery)):
    dicData = {}
    dicData["id"] = idQuery[i]
    dicData["price"] = priceScoreList[i]
    dicData["cpuSingle"] = cpuSingleScoreList[i]
    dicData["cpuMulti"] = cpuMultiScoreList[i]
    dicData["weight"] = weightScoreList[i]
    dicData["rom"] = romScoreList[i]
    dicData["screen"] = screenScoreList[i]
    dicData["hz"] = hzScoreList[i]
    dicData["power"] = powerScoreList[i]
    dicData["Xpower"] = XpowerScoreList[i]
    dicData["os"] = osScoreList[i]
    dicData["net"] = netScoreList[i]
    dicData["sim"] = simScoreList[i]
    dicData["bcam"] = bcamScoreList[i]
    dicData["fcam"] = fcamScoreList[i]
    zscoreDataList.append(dicData)
```
